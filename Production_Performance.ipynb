```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('production_data.csv')
df.head()
```

```python
df.drop('Lease', axis=1, inplace=True)
```

```python
# df.set_index('Time (Days)', inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop('Choke Size', axis=1, inplace=True)
df.head()
```

```python
df['Cum_Oil'] = df['Oil Volume (stb)'].cumsum()
df['Cum_Gas'] = df['Gas Volume (MMscf)'].cumsum()
df['Cum_Water'] = df['Water Volume  (stb)'].cumsum()
df['Cum_GI'] = df['Gas Lift Inj Volume  (MMscf)'].cumsum()
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))

A = df['Time (Days)']
B = df['Cum_Oil']
C = df['Cum_Gas']
D = df['Cum_Water']

plt.plot(A, B, color='green', linestyle='--', linewidth=3, markersize=6, label='Cum_Oil')
plt.xlabel('Time')
plt.ylabel('Volume')
plt.plot(A, C, color='red', linestyle=':', linewidth=3, markersize=6, label='Cum_Gas')

plt.plot_date(A, D, color='blue', linestyle='solid', linewidth=1, markersize=3, label='Cum_Water')

x_ticks = [day for day in A if day%365 == 0]
x_labels = ['Year ' + str(i+1) for i in range(len(x_ticks))]
plt.xticks(x_ticks,x_labels)
plt.title('Cummulative Oil, Gas and Water Production')
plt.legend(loc='upper left')
plt.show()
```

Plot the same, but on two axis (left-axis for Oil and Water, right-axis for Gas)

```python
# plt.figure(figsize=(40, 20))
plt.style.use('dark_background')
from matplotlib.ticker import StrMethodFormatter
fig, ax1 = plt.subplots(figsize=(20, 10))

A = df['Time (Days)']
B = df['Cum_Oil']
C = df['Cum_Gas']
D = df['Cum_Water']

ax1.plot(A, B, color='yellow', linestyle='--', linewidth=6, markersize=6, label='Cum_Oil')
ax1.set_xlabel('Time(Days)')
ax1.set_ylabel('Oil [STB], Water [STB]')
ax1.tick_params(axis='y', labelcolor='yellow',direction='inout')
# ax1.legend(loc='upper left')
ax1.set_title('Cummulative Oil, Gas and Water Production')

ax1.plot(A, D, color='blue', linestyle='solid', linewidth=3, markersize=3, label='Cum_Water')
ax1.legend(loc='upper left')
plt.grid(True)

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

ax2.set_ylabel('Gas [MMSCF]')
ax2.plot(A, C, color='red', linestyle=':', linewidth=6, markersize=6, label='Cum_Gas')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc='upper right')

fmtr = StrMethodFormatter(('{x:,g}'))
ax1.yaxis.set_major_formatter(fmtr)
ax2.yaxis.set_major_formatter(fmtr)


# x_ticks = [day for day in A if day%365 == 0]
# x_labels = ['Year ' + str(i+1) for i in range(len(x_ticks))]
# plt.xticks(x_ticks,x_labels)
plt.title('Cummulative Oil, Gas and Water Production')
# plt.grid(True)
plt.show()
```
